```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
```

```python
df = pd.read_csv('./Data/currencies.csv')
```

```python
df.head()
```

```python
df.set_index('Currency',inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df=df.fillna(method='ffill')
```

```python
len(df)
```

```python
df.dtypes
```

```python
df[['Colombian Peso','Korean Won','Iranian Rial','Indonesian Rupiah']].describe()
```

```python
df[['Colombian Peso','Korean Won','Iranian Rial','Indonesian Rupiah']].head()
```

```python
for col in ['Colombian Peso','Korean Won','Iranian Rial','Indonesian Rupiah']:
    print(col)
    df[col] = df[col].apply(lambda x: str(x).replace(',',''))
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype('float64')
```

```python
corr_data = df.corr()
```

```python
corr_data
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(corr_data,annot=False)
plt.show()
```
